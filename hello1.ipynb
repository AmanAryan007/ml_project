```python
import gc
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/file/hello_1.csv")
```

```python
df.dropna(inplace=True)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import pandas as pd

# Load your malware dataset into a pandas DataFrame

# List of columns with categorical values
categorical_columns = ['Sport',"Proto", 'SrcAddr', 'Sport', 'Dir', 'DstAddr', 'Dport','State','sTos','dTos']
# Loop through each categorical column and perform nominal encoding
for col in categorical_columns:
    # Use pandas' astype method to convert the column to the 'category' data type
    df[col] = df[col].astype('category')
    # Use the cat.codes attribute to replace the column values with their corresponding category codes
    df[col] = df[col].cat.codes
```

```python
# Convert 'StartTime' column to int by extracting hours, minutes, and seconds
df['StartTime'] = df['StartTime'].apply(lambda x: int(x.split(':')[0]) * 3600 + int(x.split(':')[1].split('.')[0]) * 60 + int(x.split(':')[1].split('.')[1]))
```
